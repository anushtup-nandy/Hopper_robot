## Nature of the 3D inverted Pendulum mode:

```python
#defining constants:
G=9.81

t=np.arange(-0.5,0.5,0.01)

Z_c=1
plt.figure()

C=(9.81/Z_c)
ax=plt.axes(projection='3d')

x=np.e**(C*t) + np.e**(-C*t)  #x0=2
y=np.e**(C*t) - np.e**(-C*t)   #y0=0 

for i in range(len(t)):
    ax.plot3D(x,y,Z_c)
    x1,x2=x[i],0
    y1,y2=y[i], 0
    plt.plot([x1,x2], [y1,y2], [Z_c, 0])
    plt.xlim(-5,5)
    plt.ylim(-5,5)
    plt.ylabel('y')
    plt.xlabel('x')
    plt.title('Nature of 3D LIPM')
    
plt.show()
```
